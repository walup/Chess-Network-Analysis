## Game 4 J. van der Wiel - R. Vaganian

Theme: Over-Extended center

It is important to support the center with other pieces or it could collapse. 

There is a balance between development and building a large pawn center. 



```python
from ChessGame import ChessBoard
from ChessGame import PieceColor
```

```python
board = ChessBoard()
board.initializeBoard()

board.makeMove("e4")
board.makeMove("Nf6")
board.drawBoard()
               
```

This is called the Alekhine defense. The most reasonable move for white here is e5, after which black will have to move the knight to d5 probably. 

```python
board.makeMove("e5")
board.makeMove("Nd5")
board.drawBoard()
```

As a beginner the obvious way to continue is to start chasing the knight, but maybe it could lead to the creation of some pawn weaknesses?, i don't know. 

```python
board.makeMove("d4")
board.makeMove("d6")
board.drawBoard()
```

This challenges white's center. Never let the opponent do what they want. 

```python
board.makeMove("c4")
board.drawBoard()
```

O.k. white decides to chase the knight. It can only move to either b6 or b4 from what i can see. 

```python
board.makeMove("Nb6")
board.makeMove("f4")
board.drawBoard()
```

f4 is a move that is usually considered bad, but in this case it is viable since white has gained an important ammount of space on the kingside. If we compute the space of both players we can see that white has a bit of an advantage in that area:

```python
print("white space: "+str(board.getSpace(PieceColor.WHITE)))
print("black space: "+str(board.getSpace(PieceColor.BLACK)))
```

It is still quite risky though and you have a responsability to protect all those feeble pawns. A nice move could be Be3 here. 

```python
board.makeMove("dxe5")
board.drawBoard()
```

Every time that you move a pawn something opens. Pawns are doors in this game. In this case the queen is now controlling the d-file.

```python
board.makeMove("fxe5")
board.drawBoard()
```

If white takes with the d-pawn white losses the right to castle and also black will be ahead in development. 

```python
board.makeMove("Nc6")
board.drawBoard()
```

Threatening to take on d4. 

```python
board.makeMove("Be3")
board.drawBoard()
```

Defending the pawn. Appearently here if white defends with the knight he loses tempo, since the bishop will pin it. Its kind of subtle to my liking but GMs do care about this stuff. 

```python
board.makeMove("Bf5")
board.drawBoard()
```

In the current state of the board that was the best place where black can develop his bishop. e6 is the next natural move to free its black-squared bishop, finishing his development. 

```python
board.makeMove("Nc3")
board.drawBoard()
```

d5 would have been tempting, but principles are principles, and white needs to finish his development as fast as he can. 

```python
board.makeMove("e6")
board.drawBoard()
```

Preparing the bishop to move to either e7 or d4 which would pin the knight (Temporarilly at least).

```python
board.makeMove("Nf3")
board.drawBoard()
```

```python
board.makeMove("Bg4")
board.drawBoard()
```

This threatens to take the knight, which is vital for protecting white's structure. As many GM ideas the real gain in doing this is only seen after more moves have been played. 

```python
board.makeMove("Be2")
board.drawBoard()
```

```python
board.makeMove("Bxf3")
board.drawBoard()
```

white cannot recapture with the bishop, because he will lose the pawn on c4. His only hop is to recapture with the g pawn and try to castle queenside maybe?

castling queenside would be good in my view because you can defend the d4 pawn with the queen and rook.

```python
board.makeMove("gxf3")
board.makeMove("Qd7")
board.drawBoard()
```

```python
board.makeMove("f4")
board.drawBoard()
```

Reinforcing the e5 pawn. White has moved a lot of pawns though, wouldn't it be better to just castle queenside as quick as possible?

```python
board.makeMove("Rd8")
board.drawBoard()
```

It seems like black has decided either to stay in the center on castle kingside. 

If black had castled queenside c5 allows white to break the center and attack black. In chess and in every other complex system specificity beats general rules almost always. This is why biological things are hard to deal with, we know "generally" how they should work, but exceptions matter. 

```python
board.makeMove("d5")
board.drawBoard()
```

Over-extending the center. 

The point of the game is that if a center cannot hold it is usually necessary to over-extend it, which leads to bad configurations. 

```python
board.makeMove("Bb4")
board.drawBoard()
```

```python
board.makeMove("Bxb6")
board.drawBoard()
```

This is a mistake appearently because the dark-squared piece is an important piece for white, without it the black squares in white's territory are not very well guarded. 

In chess the battle occurs in two worlds, the black squares and the white squares, and it is important not to weaken one of them, because the enemy could exploit it. 

```python
board.makeMove("axb6")
board.drawBoard()
```

```python
board.makeMove("Bf3")
board.drawBoard()
```

The lessons here are: 

<ol>
    <li>Large centers come with a price, you need to support them well. 
</li>
    <li>If you have a development advantage and you put pressure on your opponent center, they might have to over-extend it, leading to possible weaknesses.</li>
    <li>Don't forget to distribute your defenses over black and white squares. The game occurs in two worlds, and neglecting one of them can be fatal. </li>
    <li>Specificity beats general principles in chess and most complex systems.  </li>
</ol>
