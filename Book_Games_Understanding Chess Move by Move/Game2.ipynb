Queen's gambit 👑

Normal stuff. 
The purpose of this gambit is that if black accepts it by playing dxc4, then white will play e4, gaining control of the center. 

The gambit is not really much of a gambit, because it is hard to defend the pawn after it takes on c4. Black usually runs into positional troubles if they try to defend it.

```python
board.makeMove("e6")
board.drawBoard()
```

Queen's gambit declined. An alternative way of declining it is c6, which leads to the slav defense. 

```python
board.makeMove("Nf3")
board.drawBoard()
```

```python
board.makeMove("c6")
board.drawBoard()
```

Black makes a triangle of pawns. This configuration is known as the semi-slav. The reason is that it's sort of an amalgamation of the slav and the queen's gamit declined.

In queen's gambit declined you play e6 to defend the pawn on d5, in the slav you play c6 with the same purpose. In the semi-slav you end up playing both c6 and e6.

```python
board.makeMove("g3")
board.drawBoard()
```

The idea is of course to move the bishop to g2 later. This move at first glance can seem weird. Why would you want to control the h1-a8 diagonal when there are 3 pawns blocking it?

Well one of the ideas of the semi-slav is that you want to eventually play dxc4, and then b5 to protect the pawn. But now this would weaken the diagonal (only the pawn on c6 would be left). One of the drawbacks mentioned by the author is that now dxc4 would actually win a pawn, since you won't be able to to regain it with the bishop.

Defending the diagonal.

In this position, white is better positionally, while black is ahead by one pawn. One of the harder things to know as a player is what are the demands of the position. 

In this case white needs to keep his positional advantage by keeping black off-balance. The way white can lose here is by allowing black to develop and castle, regaining some steadiness.
