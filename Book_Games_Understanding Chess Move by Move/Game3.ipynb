### Game 3 K. Aseev - N.Rashkovsky

Theme of the game: controlling the center (or centre if you live in the UK xd).

Most pieces have the best mobility when they are located in the center, and until you know which side your opponent's king will castle its better to keep your configuration centralized. 

There's no simple rule to achieve center control, it might require using pawns, or sometimes other pieces. You need to recognize how it can be achieved in the most optimal manner.

```python
from ChessGame import ChessBoard
```

```python
board = ChessBoard()
board.initializeBoard()
```

```python
board.makeMove("e4")
board.makeMove("g6")
board.drawBoard()
```

g6 seems like a weird move, because it permits white to develop freely without much opposition. The idea here is to be flexible, black will wait for a little bit to find the most appropriate way to challenge white's control of the center. Usually this would involve playing something like c6 or e5 in the near future. 

g6 is known as the modern defense.

As a rule of thumb, if your opponent allows it play two pawns in the center.

The lessons of this game are. 

<ol>
    <li>Control of the center is crucial, since it gives you the flexibility to play in whichever side of the board you want.</li>
    <li>The modern defense is adaptable but you need to be careful, and not allowed your opponent to solidify the control of the center. </li>
    <li>When your opponent has no control of the center they will often have to concede positional advantages when you attack on many flanks. </li>
