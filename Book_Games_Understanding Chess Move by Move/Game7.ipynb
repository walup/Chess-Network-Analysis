### Game 7 G.Kasparov - A. Karpov

Theme: **Bring up the Reserves** 

It is common that the king will have defensive pieces and shielding pawns. You need to bring many pieces to participate in an attack if you aim to be successful. 

Set pieces in aggressive positions before commiting to an attack.
