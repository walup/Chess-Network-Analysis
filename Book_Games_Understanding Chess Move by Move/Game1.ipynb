### Game 1 A. Chernin - J. van der Wiel

```python
from ChessGame import ChessBoard
```

```python
board = ChessBoard()
board.initializeBoard()
```

```python
board.makeMove("Nf3")
board.drawBoard()
```

Nf3 is a flexible move, you're waiting to see what black will do before doing a more compromising move. This opening move can lead to transpositions to different configurations, like the English opening and the Sicilian

```python
board.makeMove("d6")
board.drawBoard()
```

```python
board.makeMove("g3")
board.drawBoard()
```

This prepares for fianchetto, which helps white to gain control of the long diagonal.

```python
board.makeMove("e5")
board.drawBoard()
```

quite a natural move. Now white finishes his fianchetto

```python
board.makeMove("Bg2")
board.drawBoard()
```

```python
board.makeMove("f5")
board.drawBoard()
```

Two many pawn moves!
The rule of thumb given by the author is: don't play more than three pawn moves in the first eight moves of the game.

In this case the pawn does achieve something. Notice that white has control of the e4 and d4 squares with his light-square bishop. The move f5 helps to regain some control of the e4 square. 

```python
board.makeMove("c4")
board.drawBoard()
```

with this move white doubles his influence on d5 (the bishop will be covering it also once the knight on f3 moves).

```python
board.makeMove("Nf6")
board.drawBoard()
```

Fighting for control of the d5 square. 

```python
board.makeMove("b4")
board.drawBoard()
```

This is a strange move for me. I think the most logical thing to do would be to contest for the d5 square again by playing Nc3.

The reasoning behind b4 is that you're setting up your attack on the queenside, so at some point you will have to play b4. However, doing this will be more difficult once there is a knight on c6. Thus, b4 is a preemptive move that might save time for white in the future. 



```python
board.makeMove("e4")
board.drawBoard()
```

This threatens the knight and gains a bit of space. The move is not so good, because now the a1-h8 diagonal is weakened. Also it is a pawn move, and while it kind of gains a tempo one has to consider that black has not developed his pieces at all.

```python
board.makeMove("Nd4")
board.makeMove("d5")
board.drawBoard()
```

5th pawn move by black!, this is not going to be good.

```python
board.makeMove("cxd5")
board.makeMove("Nxd5")
board.drawBoard()
```

```python
board.makeMove("b5")
board.drawBoard()
```

The move protects the pawn which was on b4, since it was being attacked twice. It also prevents black from developing his knight to c6 (which is its more natural square).

```python
board.makeMove("Nf4")
board.drawBoard()
```

notice that when the pawn takes the knight the queen can take the knight on d4, and white will have double pawns on the f file. Double pawns are generally undesirable, but in this case it appearently does not matter to have them. 

```python
board.makeMove("gxf4")
board.makeMove("Qxd4")
board.drawBoard()
```

```python
board.makeMove("Nc3")
board.drawBoard()
```

You won't be taking my rook today. 

```python
board.makeMove("Bc5")
board.drawBoard()
```

This threatens mate, but it can be defended quite easily. For starterst there's e3.

```python
board.makeMove("e3")
board.drawBoard()
```

```python
board.makeMove("Qc4")
board.drawBoard()
```

This prevents white from castling, but the queen is quite exposed in that position. 

Notice that the a1 -h8 diagonal is almost completely in white's control, and that black has not developed a lot of his pieces. 

```python
board.makeMove("Bb2")
board.drawBoard()
```

Taking control of the diagonal, always look for imbalances you can exploit!

```python
board.makeMove("O-O")
board.drawBoard()
```

```python
board.makeMove("Rg1")
board.drawBoard()
```

Putting pressure on the g7 pawn. This position looks horrible for black. The rook on g1 can be uncovered by attacking the queen with Bf1 gaining a tempo.

```python
board.makeMove("Rf7")
board.drawBoard()
```

Quite smart, defending the threatened pawn on g7. 

```python
board.makeMove("Rc1")
board.drawBoard()
```

Threatening to take either the queen or the bishop that are aligned in the c file. This is a common motif in games, try not to align undefended pieces in the same file or rank. 

```python
board.makeMove("Qb4")
board.drawBoard()
```

```python
board.makeMove("Qc2")
board.drawBoard()
```

Which defends the attacked bishop. Notice that on white uncovers his rook + queen battery with Nd5 white would threaten and in fact win the bishop after a few moves.  

```python
board.makeMove("Bd6")
board.drawBoard()
```

```python
board.makeMove("Nd5")
board.drawBoard()
```

The ball is rolling now. 

```python
board.makeMove("Qxb5")
board.drawBoard()
```

Losing a pawn doesn't matter when you're about to invade normandy. 

```python
board.makeMove("Nf6+")
board.drawBoard()
```

Black can not take the pawn, because then Bf1+ wins the queen.
It also cannot take with the rook, because the bishop on b2 recaptures and we sort of have the same problem. The king has no other option but to escape. 

```python
board.makeMove("Kh8")
board.drawBoard()
```

```python
board.makeMove("Bf1")
board.drawBoard()
```

The point of this move is to play Bc4, which would leave a bishop pair pointing toward the king. This would almost be imminent mate. 

```python
board.makeMove("Qxb2")
board.drawBoard()
```

Sad times we are living.

```python
board.makeMove("Qxb2")
board.drawBoard()
```

```python
board.makeMove("Rxf6")
board.drawBoard()
```

This minimizes black loss of material.

White has won a queen, black has 1 pawn + 1 knight + 1 bishop 

The count gives:

9 -1-3-3 = 2

white has +2 of material, which numerically is not a whole lot. Positionally of course white is much better, so when you introduce this thing into an engine it does not say +2, it says +5.2.

```python
board.makeMove("Qxf6")
board.drawBoard()
```

White will still be a piece up, and black's king will now be open. This leads to an endgame where white will have 2 rooks + 1 bishop, while black will have 1 rook + 1 bishop + 1 knight. (see cell 42)

```python
board.makeMove("gxf6")
board.drawBoard()
```

```python
board.makeMove("Bc4")
board.drawBoard()
```

The awaited move is now here. Notice that if black does not sacrifice his bishop he would get mated with Rg8. 

```python
board.makeMove("Be6")
board.drawBoard()
```

Sacrificing the bishop in order to protect the 8th rank with the rook. 

```python
board.makeMove("Bxe6")
board.makeMove("Na6")
board.drawBoard()
```

After sacrificing his bishop black has finally defended against being mated with the rook. 

```python
board.makeMove("Ke2")
board.drawBoard()
```

```python
board.makeMove("Nc5")
board.drawBoard()
```

Trying to defend the e4 pawn

```python
board.makeMove("Bxf5")
board.makeMove("Rd8")
board.makeMove("Rg3")
```

```python
board.drawBoard()
```

Here black resigned. 

Some reasonable continuation of the game is:


```python
board.makeMove("Rg8")
board.makeMove("Rxg8+")
board.makeMove("Kxg8")
board.makeMove("Rg1+")
board.makeMove("Kh8")
board.makeMove("f3")
board.makeMove("exf3+")
board.makeMove("Kxf3")
board.drawBoard()
```

In this position white will advance the pawns on the d and e files, and will ultimately be able to make a queen. 

The lessons of this game are:

<ol>
    <li>Do not make a lot of pawn moves in the opening. You can gain space with minor pieces like bishops, which would help you control key diagonals. </li>
    <li>There are some cases where leaving your king in the center is not a weakness </li>
    <li>Open files are critical, and can make you threaten some interesting discover checks.</li>
    <li>Be careful about your queen and the threats that could be made to her by minor pieces. </li>
    
</ol>
