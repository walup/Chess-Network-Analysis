### Game 6. G. Kasparov - N. Short

Theme: Basic attacking ideas

From what the book describes it seems that there are different elements that are important for attacks. The first is to force weaknesses, which usually leads to opening lines, having opportunities of removing defenders, and finally delivering the checkmate. 

An important concept for attacks is momentum, they require to play sequences of moves that force the opponent into the positions you want. If an attack is not continuous then the defender will have a chance to organize his position and stop the threat.
