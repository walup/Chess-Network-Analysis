Let's see how the average degree of the graph changes for black and white as a game evolves. In this case i will do it with the opera game (white wins the game).

```python
gameMoves = ["e4", "e5", "Nf3", "d6", "d4", "Bg4", "dxe5","Bxf3", "Qxf3", "dxe5","Bc4", 
             "Nf6", "Qb3",  "Qe7", "Nc3", "c6", "Bg5", "b5","Nxb5", "cxb5", "Bxb5+", 
             "Nbd7",  "O-O-O", "Rd8", "Rxd7", "Rxd7", "Rd1", "Qe6", "Bxd7+","Nxd7",
             "Qb8+","Nxb8", "Rd8#"]

whiteDegree = []
blackDegree = []

nMoves = len(gameMoves)
board = ChessBoard()
board.initializeBoard()

for i in range(0,nMoves):
    move = gameMoves[i]
    board.makeMove(move)

    whiteGraph = ChessGraph(board, True, PieceColor.WHITE)
    blackGraph = ChessGraph(board, True, PieceColor.BLACK)
    whiteDegree.append(whiteGraph.getAverageDegree())
    blackDegree.append(blackGraph.getAverageDegree())
    
```

```python
plt.figure()
plt.plot(whiteDegree, label = "White", color = "#21b5ff")
plt.plot(blackDegree, label = "Black", color = "#e87d31")
plt.xlabel("Move")
plt.ylabel("Avg. degree")
plt.legend()
```

Morphy connects the board better and ends up winning.
