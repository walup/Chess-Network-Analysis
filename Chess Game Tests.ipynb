### The opera game

Let's look at the classic opera game, which was played between Paul Morphy and two people, a duke and a count. 

I won't describe it in depth because i'm just not very good at chess. But i think Ben Finegold has a video on it, and GothamChess has also analyzed it in: https://www.youtube.com/watch?v=rzGTG04u1gs

```python
board = ChessBoard()
board.initializeBoard()
board.drawBoard()
```

O.k so black plays some sort of Phillidor defense, which is not considered very good at GM level. Us players who suck at chess have much bigger problems than the slight disadvantage that arises from this position though:

```python
board.makeMove("e4")
board.makeMove("e5")
board.makeMove("Nf3")
board.makeMove("d6")
board.drawBoard()
```

The most evident problem is that you are blocking your bishop on f8 from developing. 

The most usual response to this position is of course d4, which is exactly what Paul Morphy played. 

```python
board.makeMove("d4")
board.drawBoard()
```

Now black pins the knight with Bg4

```python
board.makeMove("Bg4")
board.drawBoard()
```

Here Morphy takes on e5, and you cannot really recapture right away, because white would take your queen and the right to castle. 

```python
board.makeMove("dxe5")
board.drawBoard()
```

You need to take on f3 before taking the pawn on e5. 

```python
board.makeMove("Bxf3")
board.drawBoard()
```

recapture with the queen and then black will take the pawn on e5

```python
board.makeMove("Qxf3")
board.drawBoard()
```

```python
board.makeMove("dxe5")
board.drawBoard()
```

Now white plays bc4 threatening checkmate (in a similar fashion to what happens in the scholar's mate). The difference with scholar configuration is that in this case the queen cannot be chased, so white loses nothing by threatening the checkmate.

```python
board.makeMove("Bc4")
board.drawBoard()
```

Here black can stop the checkmate while developing a piece (the knight), so Nf6 seems like a natural move.

```python
board.makeMove("Nf6")
board.drawBoard()
```

Now Morphy moves his queen to c3, threatening either Bxf7 or Qxb7.

```python
board.makeMove("Qb3")
board.drawBoard()
```

Black here chose to defend the pawn on f7, which is a sensible decision since that square is too close to the king. 

```python
board.makeMove("Qe7")
board.drawBoard()
```

I don't know what the analysis is here, but Morphy developed his knight to c3

```python
board.makeMove("Nc3")
board.drawBoard()
```

The duke and count played c6 which defends the pawn on b7 in a quite mischiveous way. 

```python
board.makeMove("c6")
board.drawBoard()
```

In this position we should be able to castle short, but that's not what Morphy did. He pinned the knight by playing bg5.


```python
board.makeMove("Bg5")
board.drawBoard()
```

Black tries to get aggressive with b5 and Morphy sacrifices the knight. This leaves a massive advantage in the left sector of the board. 

```python
board.makeMove("b5")
board.makeMove("Nxb5")
board.drawBoard()
```

black naturally takes and receives a check with the bishop:

```python
board.makeMove("cxb5")
board.makeMove("Bxb5+")
board.drawBoard()
```

Black here defends with the knight

```python
print(board.getPieceAtPosition("f","6").getPieceDescription())
```

```python
board.makeMove("Nbd7")
board.drawBoard()
```

Long castle

```python
board.makeMove("O-O-O")
board.drawBoard()
```

The duke and the count just concentrate all the pieces around the king. 

```python
board.makeMove("Rd8")
board.drawBoard()
```

It's really clear somthing has to be taken to continue the attack. What to take though?, Morphy took on d7

```python
board.makeMove("Rxd7")
board.drawBoard()
```

Now of course everything is going downhill for black. They have to pin their rook voluntarily by playing Rxd7

```python
board.makeMove("Rxd7")
board.drawBoard()
```

Now it can be attacked twice by the other rook, exploiting the pin. 

```python
board.makeMove("Rd1")
board.drawBoard()
```

What a pretty mess black is here. They now offer a queen trade hoping to get rid of the most powerful attacker.

```python
board.makeMove("Qe6")
board.drawBoard()
```

Forcing moves, always play forcing moves. Morphy does Bxd7+

```python
board.makeMove("Bxd7+")
board.drawBoard()
```

 You of course recapture with the knight (queen recapture is for crazy people).

```python
board.makeMove("Nxd7")
board.drawBoard()
```
